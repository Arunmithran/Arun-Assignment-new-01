```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import SpectralClustering
from itertools import cycle


dataset = pd.read_csv('Mall_Customers.csv')


X = dataset.iloc[:, [3, 4]].values


n_clusters = 5
spectral = SpectralClustering(n_clusters=n_clusters, assign_labels='discretize', random_state=42)
y_spectral = spectral.fit_predict(X)
labels = spectral.labels_


n_clusters_ = len(np.unique(labels))
print(f"Estimated number of clusters: {n_clusters_}")


plt.figure(figsize=(10, 7))
plt.clf()

colors = ["#dede00", "#377eb8", "#f781bf", "#4daf4a", "#e41a1c", "#ff7f00"]
markers = ["x", "o", "^", "s", "D", "*"]

for k, col, marker in zip(range(n_clusters_), colors, markers):
    class_members = labels == k
    cluster_center = np.mean(X[class_members], axis=0)
    
    plt.plot(X[class_members, 0], X[class_members, 1], marker, color=col, linestyle='None')
    plt.plot(cluster_center[0], cluster_center[1], marker, markerfacecolor=col, 
             markeredgecolor='k', markersize=14)

plt.title(f'Clusters identified by Spectral Clustering (Estimated clusters: {n_clusters_})')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.legend()
plt.show()
```
