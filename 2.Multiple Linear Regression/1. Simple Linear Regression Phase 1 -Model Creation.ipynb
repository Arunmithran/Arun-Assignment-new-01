```python
#pandas is a library and it contains the functions related to the database
import pandas as pd
```

```python
#Assign file path and this program name
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#Set the data and cross check for input datas
dataset
```

```python
#Assign specific name for input data without spelling errors
independent=dataset[["YEARS"]]
```

```python
#Cross verification for input data correctly assigned or not
independent
```

```python
#Assign specific name for output data without spelling errors
dependent=dataset[["SALARY"]]
```

```python
#Cross verification for output data correctly assigned or not
dependent
```

```python
#Datas split train & test purpose call to the sklearn model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
#The LinearRegression class from sklearn.linear_model is used to perform linear regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
#The fit method is used to train the model on the provided training data
regressor.fit(X_train,Y_train)
```

```python
#The coef_ attribute of the regressor object stores the coefficients (weights) of the features
weight=regressor.coef_
weight
```

```python
#The intercept_ attribute of the regressor object stores the intercept (bias) of the model. 
#This value represents the point where the regression line crosses the y-axis
bais=regressor.intercept_
bais
```

```python
#The predict method is used to make predictions based on the input data provided.
Y_pred=regressor.predict(X_test)
```

```python
#The r2_score function from sklearn.metrics is used to evaluate the performance of a regression model
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
#R-squared value of near 1 indicates perfect predictions, 0 - not better
r_score
```

```python
#It's used trained model to a file save and load it back later
import pickle
filename="finalized_model_linear.sav"
```

```python
#The pickle.dump function is used to serialize your trained model, wb - write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Loads and save the model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```

```python
#The result variable will contain the predicted value based on the input provided.
result
```
